```python
import pymongo
import pandas as pd
df = pd.read_csv('D:/ML_OPS/Vehicle Insurance Mlops Project/Vehicle_Insurance_MLOps/notebook/EasyVisa.csv')
df.head()
```

```python
# df should be converted into dict before we push it to mongodb

data = df.to_dict(orient='records')
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
DB_NAME = "Proj1"
COLLECTION_NAME = "Proj1-Data"
CONNECTION_URL = os.getenv("MONGODB_URL")
```

```python
client = pymongo.MongoClient(CONNECTION_URL)
data_base = client[DB_NAME]
collection = data_base[COLLECTION_NAME]
```

```python
# Uploading data to MongoDB
rec = collection.insert_many(data)
```

```python
# Load back data from mongodb

df = pd.DataFrame(list(collection.find()))
df.head(2)
```
